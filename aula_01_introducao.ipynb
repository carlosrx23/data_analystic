Cálculo de Media, Mediana e Moda

```python
#Lista de dados para os cálculos matemáticos
dados = [50,15,30,20,10]
print(dados)

#Cálculo da média:
soma=sum(dados) #sum = funcao para somar todos os dados de uma lista
print("Resultado somatorio: ",soma)

quantidade = len(dados) ## len obtem o tamanho da lista de dados, parecido com C
print("Quantidade: ", quantidade)

media = soma/quantidade
print("Media:  ", media)

#Calcula da mediana: qual o elemento central do conjunto de dados
#calculo da posicao do elemento central:

dados_ordenados = sorted(dados)
comp=len(dados_ordenados)
print("Dados ordenados", dados_ordenados)
print("Tamanho do vetor: ", comp)

print("Elemento da posicao 1: ", dados_ordenados[0])
print("Elemento da posicao 2: ", dados_ordenados[1])
print("Elemento da posicao 3: ", dados_ordenados[2])
print("Elemento da posicao 4: ", dados_ordenados[3])
print("Elemento da posicao 5: ", dados_ordenados[4])


#Calculo da posicao (indice) do elemento central:
#São 5 elementos na matriz
posicao=(comp - 1)//2
mediana = dados_ordenados[posicao]
print("Posicao da mediana",posicao)
print("Medina: ", mediana)


dados2 = [50,15,30,20,10,4] #com 6 elementos agora
dados_ordenados2=sorted(dados2)
print("Dados ordenados lista2:", dados_ordenados2)

mediana2 =  (dados_ordenados2[2]+dados_ordenados2[3])/2
print("Mediana2 : ", mediana2)
```
